# Importing libraries

```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import pandas as pd

%reload_ext autoreload
%autoreload 2
```

# 1. Functional Plot

```python
x_1=np.linspace(0,5,10)
y_1=x_1**2
plt.plot(x_1,y_1)
plt.title('Days Squared Chart')
plt.xlabel('Days')
plt.ylabel('Days Squared')
plt.show()
```

# 2. Print Multiple Plots

```python
plt.subplot(1,3,1)#here 1 position, 3 number of plot, 1 starting
plt.plot(x_1,y_1,'r')
plt.subplot(1,3,2)
plt.plot(x_1,y_1,'b')
plt.subplot(1,3,3)
plt.plot(x_1,y_1,'g')
```

# 3. Using Figure Objects

```python
fig_1=plt.figure(figsize=(5,4),dpi=100)
axes_1=fig_1.add_axes([0.1,0.1,0.9,0.9])
axes_1.set_xlabel('Days')
axes_1.set_ylabel('Days Squared')
axes_1.set_title('Days Squared chart')
axes_1.plot(x_1,y_1,label='x/x2')
axes_1.plot(y_1,x_1,label='x2/x')
axes_1.legend(loc=0)
#upper right:1, upper left:2, lower left:3, lower right:4


axes_2=fig_1.add_axes([0.45,0.45,0.4,0.3])
axes_2.set_xlabel('Days')
axes_2.set_ylabel('Days Squared')
axes_2.set_title('Days Squared chart')
axes_2.plot(x_1,y_1,'r')

axes_2.text(0,40, 'Message')
```

# 4. SubPlots

```python
fig_2,axes_2=plt.subplots(figsize=(8,4),nrows=1,ncols=3)
plt.tight_layout()
axes_2[0].set_title('plot 2')
axes_2[0].set_xlabel('X')
axes_2[0].set_ylabel('X squared')
axes_2[0].plot(x_1,y_1)

```

# 5. Appearance Options

```python
# Default colors( b: blue, g:green, r:red, c:cyan, m: magenta)
# y:yellow, k:black, w:white)
# color='0.75' creates a 75% gray
# you can use hexcodes color ="#eeefff"
# You can use color names foind next like this color ='burlywood'
# https://en.wikipedia.org/wiki/web_colors
# matplotlib.org/3.1.0/gallery/lines_bars_and_markers/linestyles.html
# matplotlib.org/3.3.0/api/markers_api.html

fig_3=plt.figure(figsize=(6,4))
axes_3=fig_3.add_axes([0,0,1,1])
axes_3.plot(x_1,y_1,color='navy',alpha=0.75,lw=2,ls='-.',marker='o',markersize=7,
            markerfacecolor='y',markeredgecolor='y',markeredgewidth=4)

axes_3.set_xlim([0,3])
axes_3.set_ylim([0,8])

axes_3.grid(True,color='0.6',dashes=(5,2,1,2))
axes_3.set_facecolor('#FAEBD7')
```

# 6. Save Visualization to a File

```python
fig_3.savefig('3rd_plot.png')
```
